```python
# Variables globales
points = {}
current_key = ""
keys = ['leftEye', 'rightEye', 'nose', 'mouth']

def onclick(event, img_width, img_height, img_name, save_dir):
    global current_key

    # Coordenadas relativas
    x_rel = event.x / img_width
    y_rel = event.y / img_height
    points[current_key] = {'x': round(x_rel, 4), 'y': round(y_rel, 4)}
    print(f"{current_key}: {points[current_key]}")

    next_index = keys.index(current_key) + 1
    if next_index < len(keys):
        current_key = keys[next_index]
        label.config(text=f"Haz clic en: {current_key}")
    else:
        # Guardar JSON
        json_path = os.path.join(save_dir, img_name.replace(".png", ".json"))
        with open(json_path, 'w') as f:
            json.dump(points, f, indent=2)
        print(f"Guardado: {json_path}")
        root.destroy()

def marcar_puntos(imagen_path, save_dir):
    global root, label, current_key, points

    current_key = keys[0]
    points = {}

    img = Image.open(imagen_path)
    img_width, img_height = img.size

    root = tk.Tk()
    root.title("Selecciona los puntos del filtro")

    canvas = tk.Canvas(root, width=img_width, height=img_height)
    canvas.pack()

    img_tk = tk.PhotoImage(file=imagen_path)
    canvas.create_image(0, 0, anchor=tk.NW, image=img_tk)

    label = tk.Label(root, text=f"Haz clic en: {current_key}")
    label.pack()

    img_name = os.path.basename(imagen_path)
    canvas.bind("<Button-1>", lambda e: onclick(e, img_width, img_height, img_name, save_dir))

    root.mainloop()

# Usar en carpeta
input_dir = "static\img"
output_dir = os.path.join(input_dir, "metadata")
os.makedirs(output_dir, exist_ok=True)

for fname in os.listdir(input_dir):
    if fname.endswith('.png'):
        fpath = os.path.join(input_dir, fname)
        marcar_puntos(fpath, output_dir)
```

```python
import cv2
import json
import os

# üìÅ Ruta donde est√°n las im√°genes PNG
directorio = "static\img"

# üóÇÔ∏è Carpeta para guardar los datos
salida = "datos_filtros"
os.makedirs(salida, exist_ok=True)

# üîß Puntos de anclaje a capturar
etiquetas = ["ojo_izquierdo", "ojo_derecho", "nariz", "boca"]

def capturar_puntos(imagen_path):
    puntos = []
    imagen = cv2.imread(imagen_path)
    clone = imagen.copy()
    cv2.namedWindow("Selecciona puntos")

    def click(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN and len(puntos) < len(etiquetas):
            puntos.append((x, y))
            cv2.circle(imagen, (x, y), 5, (0, 255, 0), -1)
            cv2.putText(imagen, etiquetas[len(puntos)-1], (x+5, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    cv2.setMouseCallback("Selecciona puntos", click)

    while True:
        cv2.imshow("Selecciona puntos", imagen)
        key = cv2.waitKey(1) & 0xFF
        if key == ord("r"):
            imagen = clone.copy()
            puntos = []
        elif key == ord("q") or len(puntos) == len(etiquetas):
            break

    cv2.destroyAllWindows()
    return puntos

# üîÑ Iterar sobre cada imagen PNG
for archivo in os.listdir(directorio):
    if archivo.endswith(".png"):
        path = os.path.join(directorio, archivo)
        print(f"\nüîç Procesando: {archivo}")
        anclas = capturar_puntos(path)

        datos = {
            "nombre": archivo.replace(".png", ""),
            "anchos_reales": [],  # Se puede llenar si conoces las dimensiones reales
            "anclas": dict(zip(etiquetas, anclas))
        }

        # üìù Guardar en JSON
        output_path = os.path.join(salida, f"{datos['nombre']}.json")
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(datos, f, ensure_ascii=False, indent=4)

        print(f"‚úÖ Datos guardados en {output_path}")
```
